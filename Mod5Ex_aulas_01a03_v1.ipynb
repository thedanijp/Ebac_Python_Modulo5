```python
import pandas as pd
import requests

# 1

# Carregando a base de dados
sinasc = pd.read_csv('https://raw.githubusercontent.com/thedanijp/Ebac_Python_Modulo5/main/SINASC_RO_2019.csv')

# Contando o número total de registros
print("Número total de registros:", sinasc.shape[0])

# Contando o número de registros únicos (removendo duplicados)
print("Número de registros únicos:", sinasc.drop_duplicates().shape[0])
```

```python
# 2) seu código aqui

# Contando valores faltantes por coluna
print(sinasc.isnull().sum())
```

```python
# 3) seu código aqui

# Selecionando as colunas de interesse
colunas_interesse = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']
sinasc_subset = sinasc[colunas_interesse]

# Contando valores faltantes no subset
print(sinasc_subset.isnull().sum())
```

```python
# 4) seu código aqui
# Removendo registros com Apgar5 faltante
sinasc_subset = sinasc_subset[sinasc_subset['APGAR5'].notnull()]

# Contando registros e valores faltantes após a remoção
print("Número de registros após remoção:", sinasc_subset.shape[0])
print(sinasc_subset.isnull().sum())
```

```python
# 5) seu código aqui
sinasc_subset['ESTCIVMAE'] = sinasc_subset['ESTCIVMAE'].fillna(9)
sinasc_subset['CONSULTAS'] = sinasc_subset['CONSULTAS'].fillna(9)
sinasc_subset
```

```python
# 6) Seu código aqui
# Substituindo valores faltantes em 'QTDFILVIVO' por zero

sinasc_subset['QTDFILVIVO'] = sinasc_subset['QTDFILVIVO'].fillna(0)

sinasc_subset
```

```python
# 7) seu código aqui
# Nova lista com as colunas desejadas
colunas_interesse = ['munResNome','IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5', 'PESO', 'CODANOMAL']

# Criando o novo DataFrame com as colunas selecionadas
sinasc_subset = sinasc[colunas_interesse]
sinasc_subset
```

```python
def categorizar_apgar(apgar):
    if 8 <= apgar <= 10:
        return 'normal'
    elif 6 <= apgar <= 7:
        return 'asfixia leve'
    elif 4 <= apgar <= 5:
        return 'asfixia moderada'
    else:
        return 'asfixia severa'

sinasc_subset.loc[:, 'classificacao_apgar'] = sinasc_subset['APGAR5'].apply(categorizar_apgar)

# Calculando as frequências da categorização
print(sinasc_subset['classificacao_apgar'].value_counts())
```

```python
# 9) seu código aqui
sinasc_subset = sinasc_subset.copy()

# Renomear as colunas
sinasc_subset.rename(columns={
    'IDADEMAE': 'idade_mae',
    'ESTCIVMAE': 'estado_civil_mae',
    'ESCMAE': 'escolaridade_mae',
    'QTDFILVIVO': 'quantidade_filhos_vivos',
    'GESTACAO': 'gestacao',
    'GRAVIDEZ': 'gravidez',
    'CONSULTAS': 'consultas_prenatais',
    'APGAR5': 'apgar_5',
    'munResNome': 'nome_municipio',
    'PESO': 'peso',
    'CODANOMAL': 'Código_malformacao'
}, inplace=True)
```
